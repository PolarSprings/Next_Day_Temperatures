```python
def sum_of_squares_residuals(x):
    SSx = np.sum(x**2) - (1/len(x))*((np.sum(x))**2)
    return SSx

def correlation_coefficient(x, y):
    r = np.sum((x - np.mean(x)) * (y - np.mean(y))) / \
        (sqrt((np.sum((x - np.mean(x))**2))) * sqrt((np.sum((y - np.mean(y))**2))))
    return r
```

```python
r_01 = correlation_coefficient(df.month, df.average)
r_02 = correlation_coefficient(df.day, df.average)
```

```python
def ColumnsMatch(x, y):
    x = np.array(x.columns)
    y = np.array(y.columns)
    z = np.random.randint(0, len(x))
    if len(x) != len(y):
        raise Exception
    elif x.shape != y.shape:
        raise Exception
    elif x.size != y.size:
        raise Exception
    elif x[z] != y[z]:
        raise Exception
    elif np.unique(x).all() != np.unique(y).all():
        raise Exception
    elif x[0:z].any() != y[0:z].any():
        raise Exception
    else:
        return print('the columns match')
```

```python
from sklearn.base import clone
```

```python
def FeatureImportances():
    rf_clone = clone(rf)
    rf_clone.random_state=42
    rf_clone.fit(X_train, Y_train)
    benchmark_score = rf_clone.score(X_train, Y_train)
    benchmark_oob = rf_clone.oob_score_
    
    for t in X_train.columns:
        rf_clonet = clone(rf)
        rf_clone.random_state=42
        rf_clone.fit(X_train.drop([t],1), Y_train)
        
        dropcol_score = rf_clone.score(X_train.drop([t],1), Y_train)
        dropcol_oob = rf_clone.oob_score_
        
        score_AsList.append(benchmark_score - dropcol_score)
        oob_AsList.append(benchmark_oob - dropcol_oob)
```
